```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sleep_cleaned.csv")
df.head()
df.info()
df.describe()
```

### Variables Used

- sleep_hours: Number of hours slept per day
- sleep_quality: Subjective sleep quality (1–10)
- stress_level: Subjective stress level (1–10), used as a proxy for mood


```python
#Sleep duration distribution
plt.hist(df["sleep_hours"], bins=10)
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.title("Distribution of Sleep Duration")
plt.show()
#Stress level distribution
plt.hist(df["stress_level"], bins=10)
plt.xlabel("Stress Level")
plt.ylabel("Frequency")
plt.title("Distribution of Stress Levels")
plt.show()
#Sleep vs Stress
plt.scatter(df["sleep_hours"], df["stress_level"])
plt.xlabel("Sleep Hours")
plt.ylabel("Stress Level")
plt.title("Sleep Duration vs Stress Level")
plt.show()
#Sleep Quality vs Stress
plt.scatter(df["sleep_quality"], df["stress_level"])
plt.xlabel("Sleep Quality")
plt.ylabel("Stress Level")
plt.title("Sleep Quality vs Stress Level")
plt.show()
df.corr()
df[["sleep_hours", "sleep_quality", "stress_level"]].corr()
```

### Key Observations

- Average sleep duration clusters around X hours.
- Higher sleep duration shows a decreasing trend in stress levels.
- Sleep quality has a stronger negative relationship with stress than sleep duration.
- Stress levels are more variable at lower sleep durations.
